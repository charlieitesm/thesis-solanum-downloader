## Adding the section to the dataframe

```python
sections = pd.read_csv("/content/drive/MyDrive/Tesis/Bibliografía/Secciones_Solanum.csv",
                       encoding="ISO-8859-1")

sections["Section"] = sections["Section"].apply(lambda sec: sec.split()[1].lower())

sections["Species"] = sections["Species"].apply(lambda sp: sp.split()[1].lower())

def fusion_species(series: pd.Series):
    unique_species = sorted(set(series))
    return ", ".join(unique_species)

species_by_section = sections.groupby('Section').agg({'Species':fusion_species}).reset_index()
display(species_by_section)
```

```python
def determine_section(row: pd.Series, species_by_section: pd.DataFrame) -> str:

    specimen_species = row["species"]
    
    return sections[sections["Species"] == specimen_species]["Section"].to_string(index=False).strip()


idigbio_images["section"] = idigbio_images.apply(lambda row: determine_section(row, species_by_section), axis=1)
```

```python
display(idigbio_images.isna().sum())
display(idigbio_images.head(20))
```
